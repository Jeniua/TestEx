Данные взяты с https://www.kaggle.com/datasets/antfarol/car-sale-advertisements.
Это набор данных о продаже автомобилей в 2016 году в Украине.

```python
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/faren/OneDrive/Рабочий стол/car_ad.csv')
display(data)
```

### Напишем функцию для подсчета числа автомобилей каждой марки.

```python
def count_car(row):
    car = row['car']
    
    return len(data.query(f'car == "{car}"'))
    
```

```python
data['count'] = data.apply(count_car,axis = 1)
```

```python
data
```

### Выведем автомобили, которых купили больше всего.

```python
car_pivot_table = pd.pivot_table(data, index = ['count','car'])
car_pivot_table.sort_values('count', ascending = False).head()
```

### Выведем автомобили, которых купили меньше всего. Есть много автомобилей, которых купили всего по 1-2 экземпляра, поэтому отнесем их к одной группе.
